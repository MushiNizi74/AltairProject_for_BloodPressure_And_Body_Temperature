```python
import altair as alt
import pandas as pd
```

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/00639/Maternal%20Health%20Risk%20Data%20Set.csv'
bp =pd.read_csv(url)
bp
```

```python
bp.info
```

```python
bp['RiskLevel'] = bp['RiskLevel'].replace({'low risk':1, 'mid risk': 2, 'high risk': 3})
bp['RiskLevel']= bp['RiskLevel'].astype('int')
bp.head()
```
